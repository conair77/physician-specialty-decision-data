```python
# Select relevent columns (0,1,2,4,5,9,15,16)

df = df.iloc[:,[0,1,2,4,9,15,16]]
df.head()
```

```python
def formatFix(string):
    # removes $
    string = string[1:len(string)]
    
    # Removes comma
    string = string.replace(',','')
    
    # casts string to float
    string = float(string)
    return string

print(formatFix(df.iloc[:,1].to_numpy()[0]))
```

```python
# Fix formating in currency column
salaryCol = df.iloc[:,1].to_numpy()

# Temporary array to hold values
temp = []

# Append fixed values to temp
for values in salaryCol:
    temp.append(formatFix(values))

# Replace column with fixed temp array
df['Total Salary'] = temp
df.head()
```

```python
def normalize(column):
    minimum = min(column)
    maximum = max(column)
    range = maximum - minimum
    
    # Apply normalization formula
    for i in range(len(column)):
        column[i] = (column[i] - minimum) / range
        
    return column
```

```python
# Replace Burnout N/A values with the average

df['Burnout %'].fillna((df['Burnout %']).mean(), inplace = True)
df['Match Rate'].fillna((df['Match Rate']).mean(), inplace = True)
```
